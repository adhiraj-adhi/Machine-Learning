# 2.4) Understanding our data

- Whenever we get any dataset we should try to understand it.
- Generally, to understand a dataset we perform the following tasks:  
	i. Asking Basic Questions  
	ii. EDA (Exploratory Data Analysis) - Univariate Analysis  
	iii. EDA - Multivariate Analysis  
	iv. Automation tool to do analysis - Pandas Profiler

### (i) Asking Basic Questions:
- We will be using train.csv (the titanic dataset)

- **(Q1) How big is the data?**

- **(Q3) What is the data type of cols?**

- **(Q4) Are there any missing values:**
  - We can use df.info() or, we can do df.isnull().sum() which will give a pandas series with count of no. of missing values in each column.

- **(Q5) How does the data look mathematically?**

```python
titanic.describe() # Gives details like count, mean, std, min, 25%, 50%, 75%, and max about our dataset's numerical column.
```

- **(Q6) Are there duplicate values?**

```python
titanic.duplicated().sum() # Gives the count of duplicated rows
```

- Note: Using drop_duplicates() of pandas we can remove duplicate.

- **(Q7) How is the correlation between columns?**
  - df.corr() => For any specific column we can do: df.corr()['Col_Name']
  - This gives the correlation between the numerical columns in a DataFrame.
  - Value ranges from -1 to 1:
    - Inverse relation = -1
    - No relation = 0
    - Direct proportional = 1
  - Why is it important?
    - Helps understand relation between features.
    - Helps in feature selection as we can drop highly correlated data to avoid duplication in model training.
    - Drop input column that has less affect on output column.

```python
titanic.select_dtypes(include='number').corr()['Survived']
```

- select_dtypes(include='number') filters only the numeric columns (e.g., Age, Fare, Pclass, etc.).

### (ii) EDA (Exploratory Data Analysis) using Univariate Analysis

- EDA (Exploratory Data Analysis) - Used for understanding data in and out.
- Univariate (Single Variable's) Analysis: Independent analysis of each column.
- We know that Data is of two types: Numerical and Categorical. Based on the type of column we perform the analysis.
- We will use train.csv (titanic dataset).
- Assuming we did ask all the above seven basic questions on the dataset, we can move on with EDA using Univariate Analysis. We do Univariate Analysis by plotting some graphs & visualizing and understanding our dataset.
- For train.csv:
  - Categorical Columns: Survived (0 for dead), Pclass (1 for elite class), Name (categorical with too many classes and hence not easy to analyze), Sex, SibSp (count of sibling/spouse traveling with the passenger), Parch (count of parent/children traveling with passenger), Ticket (categorical with too many classes and hence not easy to analyze), Cabin, and embarked [from where passenger onboarded (we have three values S, C, and Q)].
  - Numerical Columns: Age and Fare

**(a) Categorical Data Analysis:**

- Let us say we want to do analysis on Survived column. We can ask the question that out of total how many survived?
- For this we can use countplot (gives the count of each category in column) as: sns.countplot(df['Survived']
- Instead of using countplot we can even do:  
	df['Survived'].value_counts().plot(kind='bar')  
	df['Survived'].value_counts() -> Gives the count of each category.

```python
import seaborn as sns
sns.countplot(x='Survived', data=titanic)
```

```python
titanic['Survived'].value_counts()
```

```python
titanic['Survived'].value_counts().plot(kind='bar')
```

- We can get the percentage of each category and for that we can use Pie Chart as:

```python
titanic['Survived'].value_counts().plot(kind='pie')
```

```python
titanic['Survived'].value_counts().plot(kind='pie', autopct='%.2f') # autopct shows percentage on PieChart.
```

**(b) Numerical Data Analysis:**

- Understanding the distribution of a numerical column between min and max value in each bin (i.e. small range between min and max). For this, we can use a histogram:

```python
import matplotlib.pyplot as plt
plt.hist(titanic['Age'])  # We can even pass the number of bins here.
```

- We can have KDE plot alongwith histogram as:

```python
sns.distplot(titanic['Age'])  # In this plot, we will have probability on Y-axis.
```

- Boxplot: Gives five-number summary [i.e. min, median, max, Q1 (25th percentile), Q3 (75th percentile)]
  - Note: We know that min and max here may not be necessarily the min and max of our dataset.  
    min = Q1 - 1.5*IQR  
    max = Q3 + 1.5*IQR  
    median = 50th percentile
  - Any value beyond this min and max value is called a potential outlier.

```python
sns.boxplot(titanic['Fare'])
```

- Some calculations that we can perform on numerical data:

```python
print(titanic['Age'].min())
print(titanic['Age'].max())
print(titanic['Age'].mean())
print(titanic['Age'].std())
```

**Note: Distplot also tells us about the skewnesss of our dataset.**

### (iii) EDA using Bivariate and Multivariate Analysis:
- Here, we will understand and analyze the dependency between two columns.

```python
# Modules and dataset we will be using:
import pandas as pd
import seaborn as sns
tips = sns.load_dataset('tips')  # tips -> Data of a restaurant and it contains details like total_bill, tip, sex, isSmoker, etc. of customer.
titanic = pd.read_csv('../Datasets/train.csv')
flights = sns.load_dataset('flights') # In US Airlines, how many passengers traveled in a particular month of the year.
iris = sns.load_dataset('iris') # Our classic dataset - Species based on sepal_length, sepal_width, petal_length, and petal_width.
```

```python
tips.sample(5)
```

```python
flights.sample(5)
```

- Different possibilities in Bivariate Analysis:
  - Numerical - Numerical => Scatterplot
  - Numerical - Categorical => Bar Plot, Box Plot, Dist Plot
  - Categorical - Categorical => Heat Map, Cluster Map

**(a) Scatterplot (Numerical - Numerical):**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# Multivariate Analysis:
sns.scatterplot(x = 'total_bill', y = 'tip', data=tips, hue = 'sex')
```

```python
# Using 4 columns multivariate analysis
sns.scatterplot(tips, x = 'total_bill', y='tip', hue = 'sex', style = 'smoker')
```

```python
# Using 5 columns multivariate analysis:
sns.scatterplot(tips, x = 'total_bill', y='tip', hue = 'sex', style = 'smoker', size = 'size')  
# df['size'] is the number of people accompanied with the customer.
```

**(b) Barplot  [Numerical (Generally put on Y-axis)-Categorical (Generally put on X-axis)]:**

```python
# Suppose we want to get average age vs Pclass for our titanic dataset than:
sns.barplot(titanic, x='Pclass', y='Age')
```

```python
# Fare distribution in the different class:
sns.barplot(titanic, x='Pclass', y='Fare')
```

```python
# Multivariate Analysis (Fare distribution in different Pclass based on gender):
sns.barplot(titanic, x='Pclass', y='Fare', hue = 'Sex')
```

**(c) Boxplot (Numerical-Categorical):**

```python
sns.boxplot(titanic, x='Sex', y='Age') # Gives the boxplot for male and female.
```

```python
# Multivariate:
sns.boxplot(titanic, x='Sex', y='Age', hue='Survived') # Gives the boxplot based on male and female who survived and did not survive.
```

**(d) Distplot (Numerical-Categorical):**

```python
sns.distplot(titanic[titanic['Survived'] == 0]['Age'])
```

```python
sns.distplot(titanic[titanic['Survived'] == 1]['Age'])
```

```python
# If the above histogram is messy then we can use the hist=False parameter.
sns.distplot(titanic[titanic['Survived'] == 1]['Age'], hist=False)
```

- Analysis that we get from the above plot:-
  - For children: Probability of Surviving > Probability of Dying
  - For mid-age (18-35): It is reverse
  - And so on we can analyze

**(e) HeatMap (Categorical - Categorical):**

```python
# In each Pclass how many people survived/died:
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

- pd.crosstab():
  - It is a pandas function used to compute cross-tabulation (contingency table). It counts how often a combination of values occurs in two or more columns.
  - Syntax: pd.crosstab(index, columns) -> index i.e. raw labels.

```python
# What if we want the percentage of people who survived? We can do:
titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100  # Gives a series as output and we can even plot it as:
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar') 
# Similarly, we can do for Sex vs Survived, Embarked vs Survived, etc.
```

- Dataset has non-numeric columns like "Name", and Pandas tries to apply .mean() to all columns unless you tell it not to.
- So, we are saying only apply mean() to numeric columns by doing: df.groupby('SomeColumn').mean(numeric_only=True)

**(f) ClusterMap (Categorical-Categorical):**

```python
sns.clustermap(pd.crosstab(titanic['SibSp'], titanic['Survived'])) # Closeness between different SibSp combination with survival will pe plotted.
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived'])) # Closeness between different Parch combination with survival will pe plotted.
```

**(g) Pairplot:**
- Plots the scatterplot for each pair of numerical column and hence we are not required to do that explicitly.

```python
sns.pairplot(iris)
```

```python
# Doing Multivariate Analysis:
sns.pairplot(iris, hue='species')
```

**(h) Lineplot (Numerical-Numerical):**
- Special case of scatterplot. If we join all the points of scatterplot than it is basically the lineplot.
- We use lineplot when we have time based column on x-axis (like plot displaying number of Covid-cases with each passing day/week). Ex: We had flights dataset which had number of passengers travelled on flight in each month of a given range of years (1949-1960)
- We can get total count of passengers for each year as: df = flights.groupby('year').sum() and if we want indexing than we can do:  df = flights.groupby('year').sum().reset_index()  
- Finally, we can do: sns.lineplot(df['year'], df['passengers'])

```python
flights.groupby('year').sum(numeric_only=True)
```

```python
df = flights.groupby('year').sum(numeric_only=True).reset_index()
df
```

```python
sns.lineplot(x=df['year'], y=df['passengers'])
```

**Extras:**

```python
flights.pivot_table(values='passengers', index = 'month', columns = 'year') 
# This will give us a table giving passengers count in each month of every year.
```

```python
# Now, we can use heatmap to understand this data:
sns.heatmap(flights.pivot_table(values='passengers', index = 'month', columns = 'year'))
```

- By seeing the heatmap we can easily extract many points from our data (like we can see we have more passengers during summer). We can even go for clustermap to see which months behaved similar.
