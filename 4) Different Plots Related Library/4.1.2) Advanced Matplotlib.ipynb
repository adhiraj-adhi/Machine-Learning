### (i) Advanced options w.r.to Scatter Plot in Matplotlib

```python
# Understanding an important dataset named iris (We have downloaded it and placed in the current working directory)
iris = pd.read_csv('iris.csv')
iris # It's measurements of sepal and petal's length and width of iris flowers from three species: Iris setosa, Iris versicolor, Iris virginica
```

```python
# Understanding the correlation between sepalLength and petalLength:
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Now what if we want the different species (i.e. all the three species) of iris to be in different color on the plot?

# Step I For this is assigning a number to each species of iris:
# ================================================================
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})  # Assigning a number to each species of iris
print(iris)

# Step II: Thereafter we can use the c parameter as
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Now what if we want the different species (i.e. all three species) of the iris to be in different colors on the plot?

# Step I For this is assigning a number to each species of iris:
# ================================================================
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})  # Assigning a number to each species of iris
print(iris)

# Thereafter we can use the c parameter as
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='winter') # We can control the color by specifying theme using cmap
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

- For cmap we have options like viridis (default), summer, winter, plasma, jet, etc.

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})  # Assigning a number to each species of iris
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='winter') # We can control the color by specifying theme using cmap
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.colorbar()    # This will let us know which color represents which species
plt.show()
```

- Here, we can understand that:
  - Since blue is close to 0, it is Iris-setosa
  - Similarly green is for Iris-virginica
  - Other is for Iris-versicolor

```python
# Using the alpha parameter to set the opacity of our graph:
iris['Species'] = iris['Species'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})  
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='winter', alpha=0.4) 
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.colorbar()
plt.show()
```

### (iii) Annotations

```python
# We had a batter.csv dataset containing the batter stats of some batsman in IPL
batters = pd.read_csv('batter.csv')
batters
```

```python
# We will select first 100 batsman from this list and from these first 100 batsman we will select 25 batsman randomly:
desired_list = batters.head(100).sample(25, random_state=5)
desired_list
```

- (Q) What is random_state?
  - The random_state parameter in the sample() function of pandas is used to ensure reproducibility when selecting random samples.
  - The random_state parameter controls the random number generator used to select the samples. By fixing the random_state, the sequence of random numbers becomes deterministic, ensuring the same random samples are selected every time we run the code.
  - If we set a specific value for random_state, we (and anyone running our code) will get the same results each time the code is executed.
- (Q) Why random_state = 5 only?
  - The choice of random_state=5 (or any specific value) is arbitrary. It can be any integer we choose, as long as it remains consistent in our code to ensure reproducibility.

```python
desired_list = batters.head(100).sample(25, random_state=2)   # Data will change from the above where we had randm_state set to 5
desired_list
```

```python
# We will select first 100 batsman from this list and from these first 100 batsman we will select 25 batsman randomly:
desired_list = batters.head(100).sample(25, random_state=5)
desired_list
```

```python
# Now we have this data and we want to plot scatter plot on avg vs strike_rate then:
plt.scatter(desired_list['avg'], desired_list['strike_rate'])
```

- Here, we do not know which point represents which batsman. For this we use annotation.

```python
# Example: Let us understand an example first:
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
plt.text(1,5, 'Point 1')   # The point with coordinate (1,5) will be named 'Point 1' on the plot
plt.text(2,6, 'Point 2')   # The point with coordinate (2,6) will be named 'Point 2' on the plot
plt.text(3,7, 'Point 3', fontdict={'size':12, 'color': 'brown'})   # We can even specify the size and color of the name
```

```python
# Similarly we can perform annotation for our batsman data as well:
plt.figure(figsize=(13,7))   # Just for better visualization
plt.scatter(desired_list['avg'], desired_list['strike_rate'])

for i in range(desired_list.shape[0]):
    plt.text(desired_list['avg'].values[i], desired_list['strike_rate'].values[i], desired_list['batter'].values[i], fontdict={'size': 10})
```

```python
# What if we also want to see who has more runs on the scatter plot? -> For this we can use the size (s) parameter
plt.figure(figsize=(14,10))   # Just for better visualization
plt.scatter(desired_list['avg'], desired_list['strike_rate'], s=desired_list['runs'])  # Bigger dot means more runs

for i in range(desired_list.shape[0]):
    plt.text(desired_list['avg'].values[i], desired_list['strike_rate'].values[i], desired_list['batter'].values[i], fontdict={'size': 10})
```

### (iv) Horizontal and Vertical Lines

- There are scenarios where we want horizontal and vertical lines on the plot.
- For instance, say we want to buy only those batsman whose strike rate is greater than 130. Or say, we want to buy batsman who has strike rate greater than 130 and average greater than 32.

```python
# Example: Plotting horizontal line:
plt.figure(figsize=(14,10))   # Just for better visualization
plt.scatter(desired_list['avg'], desired_list['strike_rate'], s=desired_list['runs'])

plt.axhline(130, color='green')  # green color axhline (axis-horizontal-line) where strike rate is 130

for i in range(desired_list.shape[0]):
    plt.text(desired_list['avg'].values[i], desired_list['strike_rate'].values[i], desired_list['batter'].values[i], fontdict={'size': 10})
```

```python
# Example: Plotting vertical line:
plt.figure(figsize=(14,10))   # Just for better visualization
plt.scatter(desired_list['avg'], desired_list['strike_rate'], s=desired_list['runs'])

plt.axhline(130, color='green')  # green color axhline (axis-horizontal-line) where strike rate is 130
plt.axvline(32, color='red')  # red color axvline (axis-vertical-line) where average is 32

for i in range(desired_list.shape[0]):
    plt.text(desired_list['avg'].values[i], desired_list['strike_rate'].values[i], desired_list['batter'].values[i], fontdict={'size': 10})
```

- Only SE Marsh and Butler qualifies the criteria for avg>=32 and strike rate>=130

```python
# Example: We can even have multiple horizontal/vertical lines
plt.figure(figsize=(14,10))   # Just for better visualization
plt.scatter(desired_list['avg'], desired_list['strike_rate'], s=desired_list['runs'])

plt.axhline(130, color='green')  # green color axhline (axis-horizontal-line) where strike rate is 130
plt.axhline(140, color='red')  # green color axhline (axis-horizontal-line) where strike rate is 130
plt.axvline(32, color='green')  # red color axvline (axis-vertical-line) where average is 32

for i in range(desired_list.shape[0]):
    plt.text(desired_list['avg'].values[i], desired_list['strike_rate'].values[i], desired_list['batter'].values[i], fontdict={'size': 10})
```
