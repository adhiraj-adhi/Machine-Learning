#### Firstly we will be learning to plot the following five types of graph:
- 2D Plots
- Scatter Plots
- Bar Chart: A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.

- (Q) Why to use plt.show()?
  - Interactive vs. Non-interactive:
    - In interactive environments (e.g., Jupyter Notebook), the plots often display automatically without calling plt.show().
    - In non-interactive environments (e.g., standalone Python scripts), we must call plt.show() to see the plots as without it, the plots may not be displayed.
  - It is typically called at the end of a plotting script to visualize the plots.

- It is used in Bivariate analysis.

- If we observe we can see scatter plot is nothing but 2D line plot without line connecting the points (or in other words we can say 2D line plot is a special case of scatter plot).

```python
# Example 1: Simple function:
x = np.linspace(-10,10,50)   # We know that np.linspace(a,b,n) generates n equi-spaced numbers between a and b.

y = 10*x + np.random.randint(0,300,50)   # Here, y is of the form y=mx+c
y
```

```python
# Example 2: Using scatterplot for batter.csv dataset:
df = pd.read_csv('batter.csv')    # This is actually the dataset for players stats in IPL
df
```

```python
# Example 4: Using the s parameter with a column
# (For this we will use the tips dataset present in Seaborn which has details about bills and customers of a restaurant).
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size'])   # Based on the size column we will have the size of marker
```

- The s attribute in plt.scatter() in Matplotlib is used to specify the size of the markers in a scatter plot.
- The s attribute is especially useful for visualizing an additional dimension in scatter plots. For instance, we can view the third dimension data (say size in the 2D plot of total_bill vs tip).

```python
# In above we have tiny markers since we have fewer data i.e. less count in size for customers.
plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*25)  
```

```python
# Example 5: Plotting scatter plot using plt.plot():
plt.plot(tips['total_bill'], tips['tip'], '+')   # Here, '+'  is the marker
```

```python
# Example 6: What if we do not specify marker?
plt.plot(tips['total_bill'], tips['tip'])   # This will join all the points and give a 2D line plot
```

- plt.plot() vs plt.scatter():
  - If we use plt.plot() to plot scatterplots then we may miss some functionalities of scatterplot like s (i.e. size). Also, it will become difficult to manipulate colors and all.
  - plt.plot() is fast compared to plt.scatter() and hence can be handy when dealing with large datasets.

##### (c) Bar Chart

- When doing bivariate analysis we generally do Numerical (in general, numerical value is aggregate stuff) vs Categorical analysis. Example: Average age of children (aggregate stuff) vs the colour they like.

```python
# Example 1: Simple Bar Chart (Suppose we have two list children and colors where ith columns represent the count of children liking the ith color)
children = [10, 20, 40, 10, 30]
colors = ['red', 'blue', 'green', 'yellow', 'pink']

plt.bar(colors, children)
```

```python
# Example 4: Understanding our 'batsman_season_record.csv' dataset
df = pd.read_csv('batsman_season_record.csv')
df   # It is a dataset of 5 players run in years 2015, 2016, and 2017
```

```python
# Example 5: What if we want the plot of all five batsman run in year 2015:
plt.bar(df['batsman'], df['2015'])
```

```python
# Example 6: What if we want the plot of all five batsman runs in the year 2015 and 2016 side-by-side? One possible solution we can think of is:
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
```

- But the problem here is that the plot for year 2016 (orange color) overlaps the plot for year 2015 (blue color).
- Before understanding how to overcome this problem let us understand two things:
  - width attribute
  - np.arange() - The np.arange() function in NumPy is used to create arrays with evenly spaced values within a specified range. It is similar to Python's built-in range() function but returns a NumPy array instead of a list.

```python
# width attribute:
plt.bar(df['batsman'], df['2015'], width=0.2)
```

```python
# np.arange()
print(df.shape[0])
np.arange(df.shape[0])
```

```python
# To overcome the above problem for the side-by-side plot we do the following:
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)    # Shifting it to left by 0.2
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)    # Keeping it at it's place
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)    # Shifting it to right by 0.2
```

```python
# But now we do not have the batsman's name on the x-axis. For that we use plt.xticks() function:
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)    # Shifting it to left by 0.2
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)    # Keeping it at it's place
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)    # Shifting it to right by 0.2

plt.xticks(np.arange(df.shape[0]), df['batsman'])    # plt.xticks(what to replace, with what to replace)    
```

```python
# Here, we are having lots of information above the plot, which we can avoid using plt.show() function as:
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)    # Shifting it to left by 0.2
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)    # Keeping it at it's place
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2)    # Shifting it to right by 0.2

plt.xticks(np.arange(df.shape[0]), df['batsman'])    # plt.xticks(what to replace, with what to replace)    
plt.show()
```

##### (d) Histogram

- It is used in univariate analysis.
- For numerical column histogram is the same as bar chart is for categorical column.
- It is used for frequency count.
- Example: We have marks of students of a class. We can plot the count of students with the marks in given range (say 30-40).

```python
# Example 1: Simple example:
marks = [32, 45, 56, 10, 15, 27, 61]
plt.hist(marks)
```

- Here, our data is divided into 4 bins ([10, 22.75), [22.75, 35.5), [35.5, 48.25), [48.25, 61])

```python
# Example2: We can specify the bin count ourselves:
marks = [32, 45, 56, 10, 15, 27, 61]
plt.hist(marks, bins=5)   # Here, we will have 5 bins as ([10, 20.2), [20.2, 30.4), [30.4, 40.6), [40.6, 50.8), [50.8, 61])
```

```python
# Example 3: We can even specify beans range as:
marks = [32, 45, 56, 10, 15, 27, 61]
plt.hist(marks, bins=[0, 20, 40, 60, 80])   # Here, we have 4 bins each of size 20
```

##### (e) Pie Chart (1:46:00)

- It is used for both Univariate/Bivariate analysis.
- Can be used for Categorical or Numerical data.
- Use Case - It is used to give the contribution of a category (generally category) on a standard scale (generally scale of 100).

```python
# Example 1: Simple data:
data = [23, 45, 100, 20, 49]
plt.pie(data)
plt.show()  # just to avoid information above the graph
```

- This graph gives the contribution of each number on the total sum of data (green obviously representing the largest contribution which is of 100).
- But here we are not able to see which part in pie represents what?

```python
# Assuming our data is the marks in the i-th subject as:
data = [23, 45, 100, 20, 49]
subjects = ['eng', 'science', 'maths', 'sst', 'hindi']
plt.pie(data, labels=subjects)  # Now we will get labels
plt.show()  # just to avoid information above the graph
```

```python
# Example 2: We have a dataset named 'gayle-175.csv'. This dataset is about scores of player in a match in which Gayle scored 175:
df = pd.read_csv('gayle-175.csv')
df
```

```python
# Now, we can plot the pie chart to find out the contribution of each batsman in total score:
plt.pie(df['batsman_runs'], labels=df['batsman'])
plt.show()
```

```python
# Example 3: What if we want to see the percentage contribution of each player then we can do:
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%')
plt.show()
```

- %0.1f%%':
  - % starts the format string.
  - 0.1f:
    - 0: Specifies no padding before the decimal point.
    - .1: Indicates that the percentage should display 1 decimal place.
    - f: Means that the value should be formatted as a floating-point number.
  - %%: Escapes the % character to display it literally in the output.
  - So, each percentage will be shown with 1 decimal place, followed by %.
  - Example: If a player's contribution is 25.678%, it will display as 25.7%.

```python
# Example 4: We can specify colors of our choice as:
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', colors=['blue', 'green', 'yellow', 'pink', 'cyan', 'brown'])
plt.show()
```

```python
# Example 5: The explode parameter:
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.3, 0,0,0, 0,0.5]) # Since we have 5 batsman, we provided five values
# The first and last batsman of the dataset will be exploded
plt.show()
```

```python
# Example 6: Using the shadow attribute for shadow:
plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.3, 0,0,0, 0,0.5], shadow=True)
plt.show()
```
