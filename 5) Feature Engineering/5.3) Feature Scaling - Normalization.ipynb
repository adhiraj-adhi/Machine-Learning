# 5.3) Feature Scaling - Normalization

- It is a technique often applied as part of data preparation for machine learning.
- The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting the differences in the range of values or losing the information.
- Normalization is used on numerical columns and it transforms our data so that all values fall within a similar scale.
- Normalization basically removes the unit and bring everything into the same scale, making features unitless.
- Techniques of normalization:
  - MinMaxScaling(Mostly used)
  - Mean Normalization
  - Max absolute scaling
  - Robust scaling and many others.

### MinMaxScaling:

- Formula:  
	Xi' = (Xi - Xmin) / (Xmax - Xmin)  
	Since, Xi will lie in [Xmin, Xmax] so the output of this scaling (i.e. Xi') will always be in [0,1]

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv', header=None, usecols = [0,1,2])
''' header=None tells pandas that the CSV file has no header row, so it treats the first row as data. usecols=[0,1,2] tells pandas 
to read only columns 0, 1, and 2 from the file, ignoring the rest. '''
df.head()
```

```python
df.columns = ['Class Label', 'Alcohol', 'Malic Acid']    # providing column names
df.head()
```

```python
df.shape
```

```python
# Understanding dataset:
sns.kdeplot(df['Alcohol'])
color_dict = {1: 'red', 2: 'blue', 3: 'green'}
sns.scatterplot(x=df['Alcohol'], y=df['Malic Acid'], hue=df['Class Label'], palette=color_dict)
```

```python
# train_test_Split
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(df.drop('Class Label', axis=1), df['Class Label'], test_size=0.3, random_state = 0)
X_train.shape, x_test.shape
```

```python
# Normalization:
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(x_test)
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = x_test.columns)
```

```python
# Comparing data:
print(np.round(X_train.describe(), 1))
np.round(X_train_scaled.describe(), 1)
```

```python
# Plotting before and after scaling comparisons:
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))
ax1.scatter(X_train['Alcohol'], X_train['Malic Acid'], c = y_train)
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Alcohol'], X_train_scaled['Malic Acid'], c = y_train)
ax2.set_title("After Scaling")
plt.show()
```

### Mean Normalization & Max Absolute Scaling & RObust Scaling

- **Mean Normalization:**
  - Xi' = (Xi - Xmean) / (Xmax - Xmin)  => Making mean centric
  - Used when we want centred data. It is rarely used as instead of it people go for standardization.

- **Max Absolute Scaling:**
  - Xi' = Xi / | Xmax|
  - We have MaxAbsScaler class for this in sklearn
  - Useful when we have sparse data (i.e. data with too many zeros)

- **Robust Scaling:**
  - Xi' = (Xi - Xmedian) / IQR
  - We have RobustScaler class for this.
  - If our data has outliers than this normalization may (i.e. most of the times) perform better.

### Normalization vs Standardization:
- First question we should ask ourself: Is feature scaling required?
- If yes
	- Standardization (mostly used)
	- Normalization - Generally used when we know min and max value of our data [like in image processing (CNN)].
