- Feature Engineering is the process of using domain knowledge to extract features from raw data (i.e. preparing the data that is consumed by machine learning model from raw data -> this process is Feature Engineering).
- These features can be used to improve the performance of machine learning algorithms. In fact, if we provide well-featured data to a bad learning model then it may perform better than a good machine learning model with bad-feature-engineered data.

### Feature Engineering Classification:

**(i) Feature Transformation (Converting data from one form to another):**

- Missing Value Imputation (Removing / Filling the missing data):
  - This step is required because one of the main library, sklearn used in training the model does not accept missing values.
  - Also this is generally the first step of Feature Engineering.

- Handling Categorical Features (Encoding Categorical Data)
  - Because the sklearn library only works on numerical data. Hence, we need to convert categorical data to numerical data.

- Outlier Detection: Outlier can impact the result of machine learning model. So, they need to be removed.

- Feature Scaling
  - Generally, the last step (done just before feeding the data to the model).
  - Technique used to normalize or standardize the range of independent variables (features) in our data so they fall within a similar scale.
  - For example, if we have two features: Age in years (18 to 90) and Income in dollars (10000, 100000). Now, if we are working on some algorithms where we find magnitude or Euclidean distance between two points (like in KNN) then salary may dominate age and can skew results.

**(ii) Feature Construction (Manually creating features to make the model perform better):**
  - Creating new features (input features) from existing ones to improve the performance of machine learning model.
  - Feature Construction comes with experience or a better understanding of data or with intuition.

**(iii)Feature Selection:** Selecting only important features to train model. Example: MNIST dataset consisting of 50000 images of digit.

**(iv) Feature Extraction (Programatically extracting a new feature from a given feature):**
  - Transforms the original feature into a new set of features that still capture the important information.
  - Ex: In image pprocessing, converting a high-resolution image (thousand of pixels) into edges, colors, or textures.
  - Popular methods: PCA, LDA

- Feature Extraction vs Feature Construction:
  - Feature Extraction: Reducing original features to extract useful info.
  - Feature Construction: Creating new features from the existing ones.
  - Example: Suppose we have a dataset of real estate with features like rooms, washrooms, and price. If we want to remove feature rooms and washrooms and replace them with one i.e. by using sq. feet area. This is an example of feature extraction.
