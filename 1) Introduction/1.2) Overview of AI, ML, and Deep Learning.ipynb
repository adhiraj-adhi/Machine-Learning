#### AI

- Artificial intelligence is a branch of computer science that is concerned with **building smart and intelligent machines**.
- Examples of non-intelligent machines (i.e. machines which cannot think or make decisions): Watch, Bike, etc.
- Examples of intelligent machines: Autonomous car, Google assistant, etc.
- Evolution of AI:
  - Intelligent computer creation has been an effort for a very long time. Earlier we had **Symbolic AI** (creating knowledge system using if-else condition for each possibility).
  - Then we moved on to expert systems where the necessary knowledge for the engine was gathered from the expert of the field and then put into code. For example, chess game coded by gathering the expert's moves at a time.
  - But the problem in these systems was that we were using if-else block for each possibility. What if there is a possibility that we don't know or what if there's a fuzzy logic. For instance, say trying to know whether an image contains a dog or not.
  - To solve this problem there came machine learning.

#### Machine learning
- Machine learning is a **technique to implement AI** that can learn from the data by themselves without being explicitly programmed.
- Example: Feeding numerous images of Iron Man and Captain America to machine and asking whether a given image is of Iron Man or Captain America or none of both. Machine learning is able to determine this because it creates the pattern based on the images (i.e. data) we feed and predicts the result.

#### Deep learning
- Deep learning is a subfield of machine learning that uses **Artificial Neural Networks** to learn from the data.
- **Artificial Neural Networks:**
  - The Artificial Neural Networks is modelled from our human brain (from the neurons present in our brain).
  - Our brain has numerous neurons interconnected which makes us efficient. Similarly more the layer of neurons in deep learning more is the efficiency and prediction ability of the system.
  - Each neuron present in our brain processes the information and sends the output to another neuron. Similarly the more the layer of neurons in the deep learning the more deeper we are able to know about our data.

- Why is DL gaining popularity despite the fact that ML is powerful?
  - It tackles some of the limitations of machine learning. For instance, while training our model in machine learning we are required to specify the features of our input data (say we are required to specify features of dog and cat if we are feeding our system with images of dog and cat)
  - In deep learning our model/system automatically detects these features based on the data. This can be quite helpful when we do not exactly know the features of our data.
  - One more reason why we choose DL is that with the increase in data the performance of model increases
