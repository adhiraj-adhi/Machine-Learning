# I. Types of Machine Learning (Based on the supervision)
- Based on the amount of supervision needed machine learning can be divided into four categories:
1. Supervised Learning
2. Unsupervised Learning
3. Semisupervised Learning
4. Reinforcement Learning

#### 1. Supervised Learning

- If we have both input and output data and we want to establish relation between these input and output data so that for a new input, output can be determined. This learning is supervised machine learning.
- Example: Suppose we have data as:
  - I/P Data: IQ and CGPA of some students
  - O/P Data (often referred to as label): Whether student is placed or not
  - Now when we feed these data to machine learning then it establishes some mathematical relation between the input and output data so that it can predict the output if we provide input data of a new student.
- **Note:** Any data that we work upon is either numerical (like age, weight, cgpa, iq, etc) or categorical (like gender, nationality, company of our smartphone, etc).
- Supervised machine learning is again classified into two parts:
  - Classification: Supervised learning where our output column (i.e. output data) is categorical is classification. For instance, the above example of IQ, CGPA, and whether student is placed.
  - Regression: Supervised learning where output data is numerical data. Example, having IQ and CGPA as input and package as output data.

#### 2. Unsupervised Machine Learning
- In unsupervised machine learning we only have input data. Since we do not have output data, here task prediction is not possible (as we do not know what we have to predict).
- Therefore in unsupervised learning we do many things precisely the following four things:
  - **Clustering:**
    - It involves grouping the similar data points (input data points).
    - Real life example: Imagine we have a large collection of photos in our phone. A clustering algorithm can group similar photos together such as all our vacation photos in one group, all our work-related photos in another, and so on (without we having to manually sort them)
  - **Dimensionality Reduction:**
    - Sometimes working in supervised learning we have many input columns. These many input columns make our algorithms slow and improvement in result also minimises after certain columns. This is where dimensionality reduction comes into play.
    - It reduces the columns of input data while preserving its essential structure.
    - Dimensionality Reduction is basically a process used in data analysis and machine learning to reduce the number of variables (or features) in a data set while preserving as much information as possible.
    - Dimensionality reduction works as follows:
      - Feature selection: This method selects a subset of the most important features from the data set and discards the others. The goal is to keep only the features that contribute the most to the predicting power of the model.
      - Feature extraction: Instead of selecting a subset of existing features, feature extraction transforms the data into lower-dimensional space by creating new features (combination of original ones). The new features capture most of the variance or structure in the data. This technique reduces dimension and also helps us to visualize them in 2D and 3D space.
  - **Anomaly Detection:**
    - It identifies data points that differ significantly from the majority of the data often referred to as outliers.
    - Real-life example: In credit card fraud detection an anomaly detection system might flag a transaction as suspicious if it deviates significantly from our normal spending patterns (for example a large purchase in a different country).
  - **Association rules:**
    - It is used to discover interesting relationships or associations between different variables in a data set.
    - Real-life example: In a supermarket, association rule learning can be used to identify items that are frequently brought together. For instance, if customers often buy bread and butter together, the store might place these items next to each other to increase sales.

#### 3. Semi-Supervised Machine Learning:
- Supervised + Unsupervised Machine Learning
- Labelling (i.e. output data column) is a difficult task and requires time.
- Example: Google photos which identifies all the images that contains a person. Now what it does is it asks about that person in an image and when we label that person (say as "friend") then Google photos automatically labels all the images containing that person as "friend". No need to label each image individually. This is the core idea behind semi-supervised learning.

#### 4. Reinforcement Learning:
- Here we do not have data, instead, the agent (i.e. our model) learns from its experiences through trial and error.
- It is a type of machine learning where an agent learns to make decisions by interacting with an environment. The agent takes action, receives rewards (or penalties), and aims to maximize its cumulative reward over time. The key idea is that the agent learns from its experience through trial and error.
- For example, in chess game software:
  - Here, the chess board is the environment (environment is everything that the agent interacts with)
  - computer (our opponent) is the agent
  - computer in the environment of chess board takes actions (moves)
  - each step is rewarded with the ultimate reward being winning the game.
- Some more examples of reinforcement learning are autonomous systems like autonomous cars, automatic drones, etc.

# II. Types of Machine Learning (Based on how model is trained)

#### 1. Batch Machine Learning (a.k.a Offline ML)
- In this, we use the data as a whole to train the machine learning model i.e. there is **no incremental training** (training into pieces is called incremental training)
- Now, the data as a whole can be quite big so it is not recommended to train the model with these data in the production environment (i.e. on the server). Instead, such data are fed to the model in offline mode before production.
- Flow of training: Data ----> Feed it to Model ----> Test the Model ----> Put it in the production environment (i.e. server)
- The problem with this approach:
  - Our model is static once it is deployed in the production environment i.e. once we train it and put it in production it does not learn something new.
  - When we have some new data we pull down our model from production, retrain it with the whole data including new data, test it, and redeploy it. This cycle continues if we have new data.
  - One more problem is that since we have big data, it can be hard to process such big data and train the model.
  - There's the problem of availability also. Say there's a model that shows the news feed based on our interaction. If there is any update or important news and our model is not trained with that data then it will not be reflected in the feed.
- To tackle these problems we use online machine learning.

#### 2. Online Machine Learning
- It is an incremental training technique where we train our model with small batches (referred to as mini-batches) of data in sequential manner.
- After each training, our model gets on improving.
- **Note:** since these batches are small chunks of data, we can train our model on the server (i.e. in production) also.
- Flow of training: Some amount of data ----> Model ----> Test ----> Server (responsible to predict about new data and learn from new data) <---- Data (continuous inflow of data)
- Examples of models are: chatbots, youtube (we get video suggestions based on the video we watched just now).
- **When to use online machine learning?**
  - Whenever there is a concept drift / volatile environment (i.e. continuous change in nature of problem say in stock exchange).
  - It is cost-effective
  - Faster solution since we start with limited data and also over the period of time we train our model with small batches of data.
- We can use **sklearn** (Sci-Kit Learn) module to implement online machine learning. We can even use libraries like "River", "Vowpal Wabbit (also used in reinforcement learning)", etc.
- **Learning rate:** It determines how frequently we train over our data. We do not train our model at each incoming point as this makes our model more frequent to changes. Setting this learning rate properly is important as if not set properly our model can misbehave.
- **Out of core learning:** Sometimes we have a very huge amount of data. Training our model with such a huge amount of data as a whole is not possible then what we do is we break those data into small chunks and train our model with those small chunks using Online ML.
- **Disadvantages of Online ML:**
  - Tricky to use as it depends on many factors and variables.
  - Risky (If our model gets trained with some wrong or unwanted data). To avoid this risk we keep on monitoring our model with some technique (say like Anomaly Detection). And if our model gets trained with unwanted data then we must be in a position to roll back to the state where our model is not trained with the unwanted data.

# III. Types of Machine Learning (Based on how our model learns)
- We humans do learning in two ways:
  - Memorizing
  - Generalizing (Understanding/Extracting the principle behind something)
- Similarly in machine learning, model learns in two ways:
  - Instance Based (We can consider it corresponding to memoizing)
  - Model Based (We can consider it corresponding to generalizing)

#### Our Task:
- Whenever we learn an algorithm in future we should ask a question to ourselves as in that this algorithm is instance based or model based.
- Some of examples of model based learning are linear regression, logistic regression, decision trees, etc.
- Some examples of instance based learning are KNN, RBF, Networks, kernel function, etc.
- There are less algorithm based on instance based learning.

#### Difference between Usual/Conventional ML and Instance Based Learning
- Usual/Conventional ML:
  - We have to prepare the data to train the model.
  - Train model from training data to estimate model parameters i.e. discover patterns.
  - Generalize the rules in form of model even before scoring instances is seen.
  - Once the model is trained we can throw the data input/training data.
  - Storing models generally requires less storage.
  - Store the model in suitable form for future use.
  - Here model is used to predict for unseen scoring instance.
- Instance Based Learning
  - We have to prepare the data to train the model.
  - Do not train model. Pattern discovery postponed until scoring queue received.
  - No generalisation before scoring (i.e. query point). Only generalise for each scoring instance individually as and when seen.
  - Input or training data must be kept since each query uses part or full set of training observations.
  - Storing training data generally requires more storage.
  - There is no model to store.
  - Here training data is used to predict for unseen scoring instance.
